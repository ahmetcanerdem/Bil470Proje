# Importing pandas and reading train data.

```python
import pandas as pd
df = pd.read_csv("train.csv")
```

# Clearing data

```python
{c: df[c].dropna().values for c in df.columns}
df.dropna(how='all')
df['age_approx'] = pd.to_numeric(df['age_approx'])
```

# General Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig,ax = plt.subplots(1,2,figsize=(20,5))
sns.barplot(x=df.age_approx, y=df.diagnosis, ax=ax[0])
labels = ax[0].get_xticklabels();
ax[0].set_xticklabels(labels, rotation=90);
ax[0].set_title("Age distribution");

sns.countplot(x=df.sex ,ax=ax[1])
labels = ax[1].get_xticklabels();
ax[1].set_xticklabels(labels, rotation=90);
ax[1].set_title("Gender distribution");
```

# Age-Diagnosis Plot

```python
fig,ax = plt.subplots(1,2,figsize=(20,5))
subset = df[["age_approx", "diagnosis"]]
melan = subset[subset["diagnosis"] == "melanoma"]
others = subset[subset["diagnosis"] != "melanoma"]


sns.countplot(melan["age_approx"], color="orangered", ax=ax[0])
labels = ax[0].get_xticklabels();
ax[0].set_xticklabels(labels, rotation=90);
ax[0].set_title("Age distribution of Melanoma");

sns.countplot(others["age_approx"], color="lightseagreen", ax=ax[1])
labels = ax[1].get_xticklabels();
ax[1].set_xticklabels(labels, rotation=90);
ax[1].set_title("Age distribution of Others");
```

# Gender-Diagnosis Plot

```python
fig,ax = plt.subplots(1,2,figsize=(20,5))
subset = df[["sex", "diagnosis"]]
melan = subset[subset["diagnosis"] == "melanoma"]
others = subset[subset["diagnosis"] != "melanoma"]


sns.countplot(melan["sex"], color="orangered", ax=ax[0])
labels = ax[0].get_xticklabels();
ax[0].set_xticklabels(labels, rotation=90);
ax[0].set_title("Gender distribution of Melanoma");

sns.countplot(others["sex"], color="lightseagreen", ax=ax[1])
labels = ax[1].get_xticklabels();
ax[1].set_xticklabels(labels, rotation=90);
ax[1].set_title("Gender distribution of Others");
```

# Anatom Site-Diagnosis Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig,ax = plt.subplots(1,2,figsize=(20,5))
subset = df[["anatom_site_general_challenge", "diagnosis"]]
melan = subset[subset["diagnosis"] == "melanoma"]
others = subset[subset["diagnosis"] != "melanoma"]


sns.countplot(melan["anatom_site_general_challenge"], color="orangered", ax=ax[0])
labels = ax[0].get_xticklabels();
ax[0].set_xticklabels(labels, rotation=90);
ax[0].set_title("Anatom Site distribution of Melanoma");

sns.countplot(others["anatom_site_general_challenge"], color="lightseagreen", ax=ax[1])
labels = ax[1].get_xticklabels();
ax[1].set_xticklabels(labels, rotation=90);
ax[1].set_title("Anatom Site distribution of Others");
```

# Percentages

```python
subset = df[["anatom_site_general_challenge", "diagnosis"]]
melan = subset[subset["diagnosis"] == "melanoma"]
others = subset[subset["diagnosis"] != "melanoma"]
anatomSite = subset["anatom_site_general_challenge"].unique()
anatomSite = anatomSite[~pd.isna(anatomSite)]

PercentageMelan = []
PercentageOthers = []
head = subset[subset["anatom_site_general_challenge"] == "head/neck"]
numberOfMelanHead = (head[head["diagnosis"] == "melanoma"]).size
numberOfOthersHead = (head[head["diagnosis"] != "melanoma"]).size
PercentageMelan.append(numberOfMelanHead/melan.size*100)
PercentageOthers.append(numberOfOthersHead/others.size*100)

head = subset[subset["anatom_site_general_challenge"] == "upper extremity"]
numberOfMelanUpper = (head[head["diagnosis"] == "melanoma"]).size
numberOfOthersUpper = (head[head["diagnosis"] != "melanoma"]).size
PercentageMelan.append(numberOfMelanUpper/melan.size*100)
PercentageOthers.append(numberOfOthersUpper/others.size*100)

head = subset[subset["anatom_site_general_challenge"] == "lower extremity"]
numberOfMelanLower = (head[head["diagnosis"] == "melanoma"]).size
numberOfOthersLower = (head[head["diagnosis"] != "melanoma"]).size
PercentageMelan.append(numberOfMelanLower/melan.size*100)
PercentageOthers.append(numberOfOthersLower/others.size*100)

head = subset[subset["anatom_site_general_challenge"] == "torso"]
numberOfMelanTorso = (head[head["diagnosis"] == "melanoma"]).size
numberOfOthersTorso = (head[head["diagnosis"] != "melanoma"]).size
PercentageMelan.append(numberOfMelanTorso/melan.size*100)
PercentageOthers.append(numberOfOthersTorso/others.size*100)

head = subset[subset["anatom_site_general_challenge"] == "palms/soles"]
numberOfMelanPalms = (head[head["diagnosis"] == "melanoma"]).size
numberOfOthersPalms = (head[head["diagnosis"] != "melanoma"]).size
PercentageMelan.append(numberOfMelanPalms/melan.size*100)
PercentageOthers.append(numberOfOthersPalms/others.size*100)

head = subset[subset["anatom_site_general_challenge"] == "oral/genital"]
numberOfMelanOral = (head[head["diagnosis"] == "melanoma"]).size
numberOfOthersOral = (head[head["diagnosis"] != "melanoma"]).size
PercentageMelan.append(numberOfMelanOral/melan.size*100)
PercentageOthers.append(numberOfOthersOral/others.size*100)

fig,ax = plt.subplots(1,2,figsize=(20,5))

sns.barplot(x=anatomSite, y=PercentageMelan, ax=ax[0])
labels = ax[0].get_xticklabels();
ax[0].set_xticklabels(labels, rotation=90);
ax[0].set_title("Percentage of melanoma");

sns.barplot(x=anatomSite, y=PercentageOthers, ax=ax[1])
labels = ax[1].get_xticklabels();
ax[1].set_xticklabels(labels, rotation=90);
ax[1].set_title("Percentage of others");

```
